```python
class Bank:
    def __init__(self, AccNo, Name):
        self.a =AccNo
        self.n =Name
        self.Bal =  0

    def deposit(self, d):
        self.Bal += d 
        print("Amount deposite successfully!")
    
    def withdraw(self, w):
        if w>self.Bal:
            print("Insufficient balance")
        else:
            self.Bal -= w
            print("Amount withdraw succesfully")
    
    def Check(self):
        print(f"Your Current account balance is: {self.Bal}")

    def display(self):
        print(f"Account number: {self.a}")
        print(f"Account Holder name: {self.n}")

x = Bank(123, 'XYZ')
x.deposit(1000)
x.Check()
x.withdraw(3000)
x.withdraw(500)
x.Check()
x.display()

        
```

```python
class Base:
    def __init__(self,name, Id):
        self.n= name
        self.id = Id

    def show(self):
        print(f"Name of Id no.{self.id} is {self.n}")

class Derived(Base): #Inheritance
    def speak(self):
        print("Can speak 3 language")

obj = Base("Anjali", 2023)
obj.show()
#obj.speak() #this is not possible
oj = Derived('Urmi', 2022)
oj.show()
oj.speak()
```

```python
class Singer:
    def __init__(self, name):
        self.n = name
    def show(self):
        print("Singer name")

class Stream:
    def __init__(self, country):
        self.c = country
    def show(self):
        print("Country they belong")

class Spotify(Singer, Stream): #Multiple inheritance
    def __init__(self, name, country):
        self.n = name
        self.c = country

x = Spotify("US", "Taylor Swift")
print(x.n)
print(x.c)
x.show() #whatever is written 1st will be the priority
print(Spotify.mro()) #it will show the prirority list
```

```python
class Animal:
    def __init__(self, name, species):
        self.n = name
        self.s = species
    def show(self):
        print(f"Name: {self.n} \nSpecies: {self.s}")

class Dog(Animal): #multilevel inheritance
    def __init__(self, name, breed):
        Animal.__init__(self, name, species = "Dog")
        self.b = breed
    def show(self):
        Animal.show(self)
        print(f"Breed: {self.b}")

class GoldenRetriver(Dog): #multilevel inheritance
    def __init__(self, name, color):
        Dog.__init__(self, name, breed = "Golden Retriver")
        self.c = color
    def show(self):
        Dog.show(self)
        print(f"Color: {self.c}")

obj = GoldenRetriver("Jacky", "Black")
obj.show()
```

```python
class person:
    def __init__(self,name, age, gender):
        self.n = name
        self.a = age
        self.ad = "Bhopal"
        self.g = gender

class Std(person): #Hierarchical inheritance
    def __init__(self,name, age, gender, branch):
        super().__init__(name, age, gender) #Super keyword use to override parent methods
        self.b = branch
        print("---Student details---")
    def display(self):
        print(f"Name: {self.n}")
        print(f"Age: {self.a}")
        print(f"Address: {self.ad}")
        print(f"Gender: {self.g}")
        print(f"Branch: {self.b}")

class Faculty(person): #Hierarchical inheritance
    def __init__(self,name, age, gender, school):
        super().__init__(name, age, gender)
        self.s = school
        print("---Faculty details---")
    def display(self):
        print(f"Name: {self.n}")
        print(f"Age: {self.a}")
        print(f"Address: {self.ad}")
        print(f"Gender: {self.g}")
        print(f"School: {self.s}")

s1 = Std("Urmi", 20,"Female","Btech" )
s1.display()
f1 = Faculty("Poonkurtan", 50, "Male","SCSE")
f1.display()


```

An abstract class can be considered a blueprint for other classes. It allows you to create a set of methods that must be created within any child classes built from the abstract class.

A class that contains one or more abstract methods is called an abstract class. An abstract method is a method that has a declaration but does not have an implementation.

```python
#abstrctmthd means every sub class must redefine that method
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def Area(self):
        return 0 #pass
    
class Rec(Shape):
    def __init__(self,l,b):
        self.l = l
        self.b = b
    def Area(self):
        return self.l*self.b
class circle(Shape):
    def __init__(self,r):
        self.r = r
    def Area(self):
        return 3.14*(self.r)**2
    
r = Rec(3,4)
c = circle(4)
print(r.Area())
print(c.Area())
    
```

```python
class Employee:
    def __init__(self, name, salary) -> None:
        self.n = name
        self.s = salary
    @classmethod #they can be used as alternative constructors
    def ifStr(cls, str):
        return cls(str.split("-")[0] #represanting name
                   , int(str.split("-")[1])) #represanting Salary

e1 = Employee("Urmi", 100)
print(e1.n)
print(e1.s)

string = "Aaron-200"
e2 = Employee.ifStr(string)
print(e2.n)
print(e2.s)

```

Method overloading: Super keyword use to override parent methods 

```python
class parent:
    def hello(self):
        print("this is parent method")

class child(parent):
    def hello(self):
        print("override using Super keyword")
        super().hello()
    def hii(self):
        print("this is child method")

c = child()
c.hii()
c.hello()
```

operator overloading

```python
class Vector:
    def __init__(self, i, j , k):
        self.i = i
        self.j = j
        self.k = k
    def __str__(self): #overring str
        return f"{self.i}i + {self.j}j + {self.k}k"
    
    def __add__(self, x): #override add(+)
        return Vector(self.i + x.i, self.j + x.j, self.k + x.k)
    
v1 = Vector(1,2,3)
print(v1)
v2 = Vector(4,5,6)
print(v2)

print(v1 + v2) # add--> + and we r passing v2 as argument
print(type(v1 + v2))
```
