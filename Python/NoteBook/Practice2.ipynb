Bitwise operators ------> & | ^ ~

```python
#and
a = 10  # (1010 in binary)
b = 4   # (0100 in binary)
result = a & b #===== 0000 # Output: 0
#1*0= 0    1*1=1
print(result) 

#or
a = 10  # (1010 in binary)
b = 4   # (0100 in binary) 
result = a | b  #===== 1110 # Output: 14
# 1+0 =1  1+1=0   0+0=0
print(result)  

#Xor
a = 10  # (1010 in binary)
b = 4   # (0100 in binary)
result = a ^ b #===1110
#0 1 --> 1    0 0 --> 0    1 1--> 0
print(result)  # Output: 14

#not
a = 10  # (1010 in binary)
result = ~a #------> 2's complement
print(result)  # Output: -11

#left shift
a = 10  # (1010 in binary)
result = a << 2 #===101000
print(result)  # Output: 40

#right shift
a = 10  # (1010 in binary)
result = a >> 2 #===10
print(result)  # Output: 2

```

Logical op ----> and or not

```python
result = True and False
print(result)  # Output: False
result = True or False
print(result)  # Output: True
result = not True
print(result)  # Output: False
result = True and not False or False
print(result)  # Output: True

```

______input______

The input() function in Python is a built-in function that allows you to take user input. It reads a line from the input (usually from the user), converts it to a string, and returns it. 

```python
name = input("Enter your name: ")
print("Hello, " + name)
```

```python
#To read an integer from the user, we can use the int() function.
age1 = int(input("Enter age of father: "))
print(age1)
age2 = int(input("Enter age of son: "))
print(age2)
age_sum = age1 + age2
print("Combined age of father and son is " + str(age_sum) + " years.")
```

```python
import re

def validate_password(password):
  """Validates a password based on given criteria."""

  # Adjust the regular expression to match your specific requirements
  password_regex = r"^(?=.[a-z])(?=.[A-Z])(?=.\d)(?=.[@$!%*#])(?=.{8,})"

  if re.match(password_regex, password):
    return True
  else:
    return False

# Example usage
password = input("Enter your password: ")
if validate_password(password):
  print("Valid password")
else:
  print("Invalid password")
```

```python
def count_occurrences(string, value):
  count = 0
  for char in string:
    if char == value:
      count += 1
  return count

string = "hello world"
value = "l"
result = count_occurrences(string, value)
print(f"The value '{value}' occurs {result} times in the string '{string}'.")
```

```python
def pyramid(n):
    for i in range(n):
        print('_' * (n - i - 1) + '*' * (2 * i + 1))

pyramid(5) #range (0,4)
```

```python
def inverted_pyramid(n):
    for i in range(n, 0, -1):
        print(' ' * (n - i) + '*' * (2 * i - 1))

inverted_pyramid(5)
```

```python
def number_triangle(n):
    for i in range(1, n + 1):
        for j in range(1, i + 1):
            print(j, end=' ')
        print()

number_triangle(5)
```
