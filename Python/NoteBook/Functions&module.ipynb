---Global scope---  
variables are defined in the main body. They are visible throughout the program file.

A variable which is defined within a function is local to that function. It exists as long as the function is executing.

A variable created inside a function belongs to the local scope of that function, and can only be used inside that function.

```python
y = 100

def myfunc():
  y = 300
  print(y)#y----> 300

myfunc()
print(y)#y----> 100
```

```python
y = 100

def myfunc():
  global y #perminantaly overides the global variable
  y = 300
  print(y)#y----> 300

myfunc()
print(y)#y----> 300
```

```python
x=1
def outer():
    x=2
    def inner():
        x=3
        print(x)
    inner()
    print(x)
outer()
print(x)
```

```python
x=1
def outer():
    x=2
    def inner():
        global x
        x=3
        print(x)
    inner()
    print(x)
outer()
print(x)
```

```python
x=1
def outer():
    x=2
    def inner():
        nonlocal x
        x=3
        print(x)
    inner()
    print(x)
outer()
print(x)
```

```python
a = 1

def f(): # Uses global because there is no local 'a'
    print('a Inside f() : ', a)

def g(): # Variable 'a' is redefined as a local
    a = 2
    print('a Inside g() : ', a)

def h(): # Uses global keyword to modify global 'a'
    global a
    a = 3
    print('a Inside h() : ', a)
 
print('global a : ', a)
f() 
print('global a : ', a)
g() 
print('global a : ', a)
h() 
print('global a : ', a)
```

Recursive Functions: fuc. that call itself

```python
def fact(n):
    if (n==1 or n==0):
        return 1
    else:
        return n*fact(n-1) # func. calling itself
    
n = int(input("Enter ano.:"))
print(f"The factional of num {n} is ", fact(n))
```

Decorators is fuction that takes another func as an argument

```python
def greet(fx):
    def mfx(*args, **kwargs): #args for tuple, kwargs for dic
        print("Hello!")
        fx(*args, **kwargs)
        print("Thanks for using this function!")
    return mfx

@greet
def add(a,b):
    print(a+b)

# greet(add)(3,5)
add(3,5)
```

Every Python program is a module, that is, every file that you save as .py extension is a module.

import module_name

from module-name import function-name (for Specific Attributes)

from module_name import * (for importing all)

```python
from math import sqrt, factorial

print(sqrt(16))
print(factorial(6))
```

```python
from math import *
 
# if we simply do "import math", then math.sqrt(16) and math.factorial() are required.
print(sqrt(16))
print(factorial(6))
```

```python
import math as mt
# if we simply do "import math", then math.sqrt(16) and math.factorial() are required.
print(mt.sqrt(16))
print(mt.factorial(6))
print(mt.degrees(2))   
print(mt.radians(60))  
print(mt.sin(2))
print(mt.cos(0.5))
print(mt.tan(0.23)) 
print(mt.factorial(4))
```
