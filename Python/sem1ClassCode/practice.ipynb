```python
num=int(input("Enter a number:"))
sum=0
while(num>0):
   reminder=num%10
   sum=sum+reminder
   num=num//10
print("The sum of all digit of given number is :",sum)
```

```python
num=int(input("Enter a number:"))
reverse=0
while(num>0):
  digit=num%10
  reverse=reverse*10+digit
  num=num//10
print("The reverse of given num is :",reverse)
```
