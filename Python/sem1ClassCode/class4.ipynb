using addition and substraction

```python
def swap(a,b):
    a=a+b
    b=a-b
    a=a-b
    return a,b

print(swap(10,20))
```

using mul. n div.

```python
def swap(a,b):
    a=a*b
    b=a/b
    a=a/b
    return a,b

print(swap(10,20))
```
